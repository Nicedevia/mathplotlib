```python
# pre requis
import numpy as np
import matplotlib.pyplot as plt
```

1. Tracer un graphe linéaire des points (x, y) où x = [0, 1, 2, 3, 4] et y = [1, 4, 3, 8, 5]

```python
x = [0, 1, 2, 3, 4]
y = [1, 4, 3, 8, 5] 
# sert a la taille de la figure
plt.figure()
plt.plot(x,y, label='y pour x', lw=2, c='blue',ls='--' )
# affiche les legendes des courbes
plt.legend()
# nom des axes
plt.xlabel('x')
plt.ylabel('y')
# tire
plt.title(' figure de exo 1')

# pas utile dans note book
plt.show()
#sert a sauve la figure
#plt.savefig('figure1.png')
```

fonction subplot
sert a generer une grille de differnt graphique sur la ligure 

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, c='blue')
plt.subplot(2, 1, 2)
plt.plot(x, y, c='red')
```

2. Tracer un histogramme à partir de données aléatoires générées par NumPy.

```python
# valeur gausienne
hitog = np.random.randn(100) 
# Tracer l'histogramme
plt.hist(hitog, bins=20, color='blue')  
# bins = nombre de barres dans l'histogramme
plt.title('Histogramme de données aléatoires')
plt.xlabel('valeurs')
plt.ylabel('nombre')
# affiche la grille
plt.grid(True)
```

3. Tracer un diagramme à barres à partir des données suivantes : 
categories = ['A', 'B', 'C', 'D', 'E'], valeurs = [25, 30, 15, 20, 10]

```python
categories = ['A', 'B', 'C', 'D', 'E']
valeurs = [25, 30, 15, 20, 10]

# Tracer le diagramme à barres
plt.bar(categories, valeurs, color='blue')
plt.xlabel('categories')
plt.ylabel('valeurs')
plt.title('giagramme')
```

4. Tracer un nuage de points à partir de données aléatoires générées par NumPy.

```python
x = np.random.randn(50) 
y = np.random.randn(50)
```

```python
# scatter sert au nuage de point
plt.scatter(x, y, color='blue', alpha=1)
# alpha est le bruit
plt.xlabel('x')
plt.ylabel('y')
plt.title('nuage de points')
plt.grid(True)
```

5. Tracer une courbe sinusoïdale en utilisant NumPy.

```python
x = np.linspace(0, 1* np.pi, 50)
# nb de point 10* 0.1 (pas)
y = np.sin(x)

plt.plot(x, y, color='red', label='sin(x)')

plt.xlabel('x)')
plt.ylabel('y')
plt.title('courbe sinusoïdale')
plt.legend()
plt.grid(True)
```

6. Tracer deux courbes sur le même graphe à partir des données suivantes :
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, color='red', label='sin(x)')
plt.plot(x, y2, color='green', label='cos(x)')

plt.xlabel('x)')
plt.ylabel('y')
```

7. Tracer un diagramme en secteurs (camembert) à partir des données suivantes :
categories = ['A', 'B', 'C', 'D'], valeurs = [30, 25, 15, 20]

```python
categories = ['A', 'B', 'C', 'D']
valeurs = [30, 25, 15, 20]
# defire une taille
plt.figure(figsize=(6, 6))
#autopct sert a afficher le pourcentage 
plt.pie(valeurs, labels=categories, autopct='%1.1f%%')
```

8. Tracer une courbe polynomiale à partir des données suivantes :
x = np.linspace(-5, 5, 100)
y = 3 * x**3 + 2 * x**2 - 8 * x + 5

```python
x = np.linspace(-5, 5, 100)
y = 3 * x**3 + 2 * x**2 - 8 * x + 5
```

```python
plt.plot(x, y, color='purple', label='y = 3*x^3 + 2*x^2 - 8*x + 5')
plt.xlabel('x')
plt.ylabel('y')
plt.title('courbee')
plt.legend()
plt.grid(True)
```

9. Tracer un diagramme en barres empilées à partir des données suivantes :
categories = ['A', 'B', 'C', 'D']
valeurs1 = [10, 15, 20, 25], valeurs2 = [5, 10, 15, 20]

```python
categories = ['A', 'B', 'C', 'D']
valeurs1 = [10, 15, 20, 25]
valeurs2 = [5, 10, 15, 20] 

plt.bar(categories, valeurs1, label='Valeurs 1', color='blue')
plt.bar(categories, valeurs2, label='Valeurs 2', bottom=valeurs1, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('diagramme ')
plt.legend()
```

10. Tracer un graphique en boîte (boxplot) à partir des données suivantes :
data = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(-3, 2, 100)]

```python
data = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(-3, 2, 100)]
plt.boxplot(data, labels=['A', 'B', 'C'])
#boxplot =grafique en boite

plt.xlabel('x')
plt.ylabel('y')
plt.title('ggraphique')
```
